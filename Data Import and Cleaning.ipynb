```python
# Rastgele 1 kullanıcının verilerini getir
random_user = df['user_id'].sample(1).values[0]
print(f"\nSeçilen Kullanıcı ({random_user}) Verileri:")
print(df[df['user_id'] == random_user].sort_values('event_date'))
```

```python
# 1. Aynı kullanıcı aynı gün için mükerrer kaydedilmişse sil
df = df.drop_duplicates()

# 2. Kritik sayısal kolonlardaki NaN (boş) değerleri 0 ile doldur
fill_cols = ['total_session_duration', 'victory_count', 'defeat_count', 'iap_revenue', 'ad_revenue']
df[fill_cols] = df[fill_cols].fillna(0)

print(f"Temizlik sonrası satır sayısı: {len(df)}")
```

```python
# Sadece mantıklı sınırlar içindeki veriyi alalım (Örn: 12 saatten fazla oynamayanlar)
# 12 saat = 43.200 saniye
df_clean = df[df['total_session_duration'] < 43200].copy()

# Gelir tarafında da negatif değer varsa temizleyelim
df_clean = df_clean[(df_clean['iap_revenue'] >= 0) & (df_clean['ad_revenue'] >= 0)]
```

```python
# Kullanıcı başına özet tablo çıkaralım (Bu daha sağlıklı sonuç verir)
user_summary = df_clean.groupby('user_id').agg({
    'total_session_duration': 'sum',
    'victory_count': 'sum',
    'defeat_count': 'sum',
    'iap_revenue': 'sum'
}).reset_index()

# Kazanma Oranı (Win Rate) hesapla
user_summary['win_rate'] = user_summary['victory_count'] / (user_summary['victory_count'] + user_summary['defeat_count'])
# Win rate NaN olanları (hiç maç yapmayanlar) temizle
user_summary = user_summary.dropna(subset=['win_rate'])
```
